```python
#None is Like NULL
answer = None
x = randn()
if x > 1 :
    answer = "Greater than 1"
print(x)
print(answer)
```

```python
answer = None
x = randn()
if x > 1 :
    answer = "Greater than 1"
else:
    answer = "Less than 1"
print(x)
print(answer)
```

```python
# Nested
answer = None
x = randn()
if x > 1 :
    answer = "Greater than 1"
else:
    if x >= -1 :
        answer = "Between -1 and 1"
    else:
        answer = "Less than -1"
print(x)
print(answer)
```

```python
# Chained Stmt
answer = None
x = randn()
if x > 1 :
    answer = "Greater than 1"
elif x >= -1 :
    answer = "Between -1 and 1"
else:
    answer = "Less than -1"
print(x)
print(answer)
```
